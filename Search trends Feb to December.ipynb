```python
 # Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

feb_data = pd.read_csv("febdata.csv")
mar_data = pd.read_csv("mardata.csv")
april_data = pd.read_csv("aprildata.csv", skiprows = 28)
may_data = pd.read_csv("maydata.csv", skiprows = 28)
june_data = pd.read_csv("junedata.csv", skiprows = 28)
july_data = pd.read_csv("juldata.csv", skiprows = 28)
aug_data = pd.read_csv("augdata.csv", skiprows = 28)
sep_data = pd.read_csv("sepdata.csv", skiprows = 28)
nov_data = pd.read_csv("novdata.csv", skiprows = 28)
dec_data = pd.read_csv("decdata.csv", skiprows = 28)
```

```python
#search patterns
search_patterns = [len(feb_data),len(mar_data),len(april_data),len(may_data),len(june_data),len(july_data), len(aug_data), len(sep_data), len(nov_data), len(dec_data)]
months = ['February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'November', 'December']
plt.bar(months, search_patterns, color= 'c')
plt.xticks(rotation = '30')
plt.xlabel("Months")
plt.ylabel("Number of Rising Searches")
plt.title("Number of Rising Google Searches per Month during Covid")
plt.savefig("Rising_Searches")
plt.show()
```
