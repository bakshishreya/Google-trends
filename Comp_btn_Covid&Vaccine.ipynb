```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
 # Store filepath in a variable
file_one = "Worldwide_data.csv"
file_two = "US_data.csv"
```

```python
# Read our Data file with the pandas library
# Not every CSV requires an encoding, but be aware this can come up
file_one_df = pd.read_csv(file_one, encoding="ISO-8859-1")
file_two_df = pd.read_csv(file_two, encoding="ISO-8859-1")
```

```python
#file_one_df_2=dropna(COVID-19 vaccine: (Worldwide))
```

```python
 # Show just the header
file_one_df.head(10)
```

```python
#ploting the graph for Worldwide data 
plt.figure(figsize=(30,10))
keywords=['Covid-19: (Worldwide)','Vaccine: (Worldwide))']
xaxis=file_one_df['Week']
yaxis=file_one_df['Covid-19: (Worldwide)']
yaxis2=file_one_df['vaccine: (Worldwide)']
plt.plot(xaxis, yaxis2,'-o', color="red",linewidth=5,markersize=12)
plt.plot(xaxis, yaxis,'-*', color="blue",linewidth=5,markersize=20)
#plt.bar(xaxis, yaxis, alpha=.50, color="b", align="center",)
#plt.bar(xaxis,yaxis2, alpha=.50, color="y", align="center",)
plt.title('Worldwide Comparison of Google trends between Covid-19 and Vaccine for past 10 months',fontsize =30)
plt.legend(keywords ,loc='upper right')
plt.xlabel("Week",fontsize =30)
plt.ylabel("Interest Index",fontsize =30)
plt.xticks(rotation="vertical")
plt.savefig("Worldwide_search")
plt.tight_layout()
plt.grid()
plt.show()
```

```python
#ploting the graph for US data 
plt.figure(figsize=(30,10))
keywords=['Covid-19: (United States)','Vaccine: (United States)']
x_axis=file_two_df['Week']
y_axis=file_two_df['Covid-19: (United States)']
y_axis_2=file_two_df['vaccine: (United States)']
plt.plot(x_axis, y_axis,'-o', color="g",linewidth=5,markersize=12)
plt.plot(x_axis, y_axis_2,'-*', color="y",linewidth=5,markersize=20)
#plt.bar(x_axis, y_axis, alpha=.50, color="b", align="center",)
#plt.bar(x_axis, y_axis_2, alpha=.50, color="y", align="center",)
plt.title('US Comparison of Google trends between Covid-19 and Vaccine for past 10 months',fontsize =35)
plt.legend(keywords ,loc='upper right')
plt.xlabel("Week",fontsize =30)
plt.ylabel("Interest Index",fontsize =30)
plt.xticks(rotation="vertical")
plt.savefig("US_data")
plt.tight_layout()
plt.grid()
plt.show()
```
