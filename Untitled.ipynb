```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
 # Store filepath in a variable
file_one = "multiTimeline (6).csv"
file_two = "multiTimeline (7).csv"
```

```python
# Read our Data file with the pandas library
# Not every CSV requires an encoding, but be aware this can come up
file_one_df = pd.read_csv(file_one, encoding="ISO-8859-1")
file_two_df = pd.read_csv(file_two, encoding="ISO-8859-1")
```

```python
 # Show just the header
file_one_df.head(10)
```

```python
plt.figure(figsize=(30,10))
keywords=['COVID-19 vaccine: (Worldwide)','Coronavirus diseases: (Worldwide)']
xaxis=file_one_df['Week']
yaxis=file_one_df['Coronavirus diseases: (Worldwide)']
yaxis2=file_one_df['COVID-19 vaccine: (Worldwide)']
# plt.plot(xaxis, yaxis2,'-o', color="red")
# plt.plot(xaxis, yaxis,'-*', color="blue")
plt.bar(xaxis, yaxis, alpha=.50, color="b", align="center",)
plt.bar(xaxis,yaxis2, alpha=.50, color="y", align="center",)
plt.title('Worldwide Comparision of Google trends between Covid and Vaccine for past 12 months',fontsize =20)
plt.legend(keywords ,loc='upper right')
plt.xlabel("Search per Week",fontsize =15)
plt.ylabel("Number of searches",fontsize =15)
plt.xticks(rotation="vertical")
plt.show()
```
